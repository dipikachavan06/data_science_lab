```python
def display_multiplication_table(num, limit=10):
    print(f"Multiplication Table for {num}:")
    for i in range(1, limit + 1):
        print(f"{num} x {i} = {num * i}")

# Example usage:
number_to_table = 7
display_multiplication_table(number_to_table)

print("\n") # Add a newline for better readability between examples

number_to_table_2 = 12
display_multiplication_table(number_to_table_2, limit=5) # Displaying up to 5 for the second example
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")

number = 0
result = factorial(number)
print(f"The factorial of {number} is {result}")

number = 7
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        print(f"The number {number} is Even.")
    else:
        print(f"The number {number} is Odd.")

# Example usage:
check_odd_even(4)
check_odd_even(7)
check_odd_even(0)
check_odd_even(15)
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    s_number = str(number)
    # Compare the original string with its reversed version
    if s_number == s_number[::-1]:
        print(f"The number {number} is a palindrome.")
    else:
        print(f"The number {number} is not a palindrome.")

# Example usage:
is_palindrome(121)
is_palindrome(12321)
is_palindrome(12345)
is_palindrome(0)
is_palindrome(1001)
```

## Define Expert System Concept

Explain the fundamental concepts of an expert system, including its components like knowledge base, inference engine, and user interface.


### What is an Expert System?

An **Expert System** is a computer program that emulates the decision-making ability of a human expert. It is designed to solve complex problems and provide decision support in a particular domain by reasoning through bodies of knowledge, represented primarily as if-then rules rather than conventional procedural code.

### Components of an Expert System:

1.  **Knowledge Base**:
    *   **Purpose**: The knowledge base serves as the system's long-term memory, storing all the domain-specific knowledge required to solve problems. It's akin to the expert's understanding, experience, and factual information.
    *   **Contents**: It typically contains:
        *   **Facts**: Declarative information about the domain (e.g., "If a patient has a fever, then...").
        *   **Rules**: Heuristics, associations, and relationships often expressed as IF-THEN statements (e.g., "IF the engine does not start AND the battery is dead THEN recharge the battery").
        *   **Concepts**: Definitions and relationships between domain objects.
        *   **Procedures**: Step-by-step instructions for specific tasks.

2.  **Inference Engine**:
    *   **Purpose**: The inference engine is the "brain" of the expert system. Its primary function is to interpret and manipulate the knowledge stored in the knowledge base to draw conclusions, make recommendations, or find solutions to user queries.
    *   **Function**: It employs reasoning strategies, such as:
        *   **Forward Chaining**: Data-driven approach, starting with known facts and deriving conclusions.
        *   **Backward Chaining**: Goal-driven approach, starting with a goal and attempting to prove it by finding supporting facts.
        *   **Pattern Matching**: Identifying patterns in data to match them with rules.

3.  **User Interface**:
    *   **Purpose**: The user interface (UI) is the component that facilitates communication between the user and the expert system. It enables users to input problems, ask questions, and receive solutions or explanations.
    *   **Role**: It ensures user-friendliness and often provides:
        *   **Input Mechanisms**: For users to enter data or pose questions.
        *   **Output Displays**: To present the system's conclusions, advice, or explanations in an understandable format.
        *   **Explanation Facilities**: To clarify the reasoning process the system used to arrive at a particular conclusion, enhancing user trust and understanding.

## Implement Expert System Core

### Subtask:
Write Python code to implement the expert system. This will include defining a structure for the knowledge base (rules) and creating a basic inference engine to process user input against these rules to derive a decision.


**Reasoning**:
I will define the knowledge base as a list of dictionaries, each representing an 'if-then' rule with 'conditions' and 'recommendation'. Then, I will create the `inference_engine` function to process user facts against these rules to derive a decision, as described in the instructions.



```python
rules = [
    {
        'conditions': {'weather': 'sunny', 'temperature': 'hot'},
        'recommendation': 'Go to the beach and swim!'
    },
    {
        'conditions': {'weather': 'rainy', 'temperature': 'cold'},
        'recommendation': 'Stay indoors and read a book.'
    },
    {
        'conditions': {'weather': 'sunny', 'temperature': 'mild'},
        'recommendation': 'Go for a walk in the park.'
    },
    {
        'conditions': {'weather': 'cloudy', 'windy': 'yes'},
        'recommendation': 'Fly a kite if it\'s not raining.'
    },
    {
        'conditions': {'weather': 'snowy'},
        'recommendation': 'Build a snowman!'
    }
]

def inference_engine(facts, rules):
    """
    Processes user facts against a list of rules to derive a recommendation.

    Args:
        facts (dict): A dictionary representing the current conditions (e.g., {'weather': 'sunny', 'temperature': 'hot'}).
        rules (list): A list of rule dictionaries, each with 'conditions' and 'recommendation'.

    Returns:
        str: A recommendation based on matching rules, or a default message if no rule matches.
    """
    for rule in rules:
        all_conditions_met = True
        for condition_key, condition_value in rule['conditions'].items():
            # Check if the fact exists and matches the condition
            if condition_key not in facts or facts[condition_key] != condition_value:
                all_conditions_met = False
                break # Break from inner loop if a condition is not met

        if all_conditions_met:
            return rule['recommendation'] # Return the recommendation of the first matching rule

    return 'No specific recommendation based on current conditions.' # Default message if no rule matches

# Example usage:
print("--- Expert System Demonstrations ---")

# Test case 1: Sunny and Hot
current_facts_1 = {'weather': 'sunny', 'temperature': 'hot', 'humidity': 'low'}
recommendation_1 = inference_engine(current_facts_1, rules)
print(f"Current facts: {current_facts_1}")
print(f"Recommendation: {recommendation_1}\n")

# Test case 2: Rainy and Cold
current_facts_2 = {'weather': 'rainy', 'temperature': 'cold'}
recommendation_2 = inference_engine(current_facts_2, rules)
print(f"Current facts: {current_facts_2}")
print(f"Recommendation: {recommendation_2}\n")

# Test case 3: Sunny and Mild
current_facts_3 = {'weather': 'sunny', 'temperature': 'mild', 'time_of_day': 'afternoon'}
recommendation_3 = inference_engine(current_facts_3, rules)
print(f"Current facts: {current_facts_3}")
print(f"Recommendation: {recommendation_3}\n")

# Test case 4: Cloudy and Windy (partially matching)
current_facts_4 = {'weather': 'cloudy', 'windy': 'yes', 'temperature': 'chilly'}
recommendation_4 = inference_engine(current_facts_4, rules)
print(f"Current facts: {current_facts_4}")
print(f"Recommendation: {recommendation_4}\n")

# Test case 5: Snowy
current_facts_5 = {'weather': 'snowy', 'temperature': 'freezing'}
recommendation_5 = inference_engine(current_facts_5, rules)
print(f"Current facts: {current_facts_5}")
print(f"Recommendation: {recommendation_5}\n")

# Test case 6: No match
current_facts_6 = {'weather': 'foggy', 'temperature': 'cool'}
recommendation_6 = inference_engine(current_facts_6, rules)
print(f"Current facts: {current_facts_6}")
print(f"Recommendation: {recommendation_6}\n")
```
